##**Label Encoder**

Label Encoder adalah salah satu teknik transformasi data yang digunakan dalam analisis data dan pembelajaran mesin. Teknik ini digunakan untuk mengubah variabel kategori atau label dalam bentuk teks menjadi angka, sehingga komputer dapat lebih mudah memahami dan memprosesnya. Ini adalah langkah umum dalam pra-pemrosesan data sebelum digunakan dalam model pembelajaran mesin.


Label Encoding: $x_i = \text{Angka ke-}i$

Di sini, $x_i$ adalah label atau nilai kategori ke-$i$, dan hasil Label Encoding adalah angka ke-$i$. Label Encoding mengasosiasikan setiap nilai kategori dengan nomor yang unik sesuai urutan kemunculan dalam data.

Label encoding pada data car evaluation dilakukan untuk mengubah nilai-nilai kategori atau atribut non-numerik menjadi nilai numerik. Ini dilakukan karena sebagian besar algoritma machine learning memerlukan input dalam bentuk numerik. Oleh karena itu, label encoding membantu mengkonversi kategori atau label pada data evaluasi mobil ke dalam format numerik agar dapat digunakan oleh model machine learning.

Contoh penggunaan label encoding pada data car evaluation adalah pada atribut kelas, seperti "unacc" (tidak dapat diterima), "acc" (dapat diterima), "good" (baik), dan "vgood" (sangat baik). Label encoding akan menggantikan setiap nilai kategori ini dengan nilai numerik tertentu, misalnya, 0 untuk "unacc", 1 untuk "acc", 2 untuk "good", dan 3 untuk "vgood". Ini memungkinkan model machine learning untuk bekerja dengan lebih baik karena dapat memproses dan memahami hubungan matematis antara nilai-nilai ini.

Proses label encoding sangat penting ketika kita menggunakan algoritma machine learning seperti decision trees, support vector machines, atau neural networks, yang membutuhkan data input dalam bentuk numerik. Dengan melakukan label encoding, kita dapat mengoptimalkan kinerja model dan memungkinkan model untuk mengenali pola atau hubungan dalam data car evaluation tersebut.

Korelasi adalah ukuran statistik yang digunakan untuk mengukur sejauh mana dua variabel terkait satu sama lain. Dalam analisis statistik, korelasi mengukur tingkat hubungan atau hubungan antara dua variabel. Ini adalah alat yang berguna untuk memahami apakah ada ketergantungan antara dua variabel dan dalam arah apa hubungan tersebut. Korelasi sering digunakan untuk menganalisis data dalam berbagai konteks, termasuk ilmu sosial, ilmu ekonomi, ilmu alam, dan analisis data umum.

Rumus LaTeX untuk mengukur korelasi Pearson (korelasi linear) antara dua variabel $X$ dan $Y$ adalah sebagai berikut:

$ \rho(X,Y) = \frac{\text{cov}(X,Y)}{\sigma_X \sigma_Y} $

Penjelasan:

- $\rho(X,Y)$ adalah koefisien korelasi Pearson antara variabel $X$ dan $Y$.
- $\text{cov}(X,Y)$ adalah kovariansi antara $X$ dan $Y$.
- $\sigma_X$ adalah deviasi standar (standar deviasi) dari $X$.
- $\sigma_Y$ adalah deviasi standar (standar deviasi) dari $Y$.

Koefisien korelasi Pearson berkisar antara -1 dan 1:

- Jika $\rho = 1$, ini menunjukkan hubungan linear positif sempurna antara $X$ dan $Y$.
- Jika $\rho = -1$, ini menunjukkan hubungan linear negatif sempurna antara $X$ dan $Y$.
- Jika $\rho = 0$, ini menunjukkan tidak ada hubungan linear antara $X$ dan $Y.

# **Confusion Matirx**

Confusion matrix adalah tabel yang digunakan untuk mengevaluasi kinerja model klasifikasi dengan membandingkan hasil prediksi model dengan nilai sebenarnya. Terdapat empat bagian utama dalam confusion matrix:

- True Positive (TP): Jumlah sampel positif yang benar-benar diprediksi dengan benar.

- True Negative (TN): Jumlah sampel negatif yang benar-benar diprediksi dengan benar.

- False Positive (FP): Jumlah sampel negatif yang salah diprediksi sebagai positif (kesalahan Type I).

- False Negative (FN): Jumlah sampel positif yang salah diprediksi sebagai negatif (kesalahan Type II).

Rumus Confusion Matrix :

\begin{bmatrix}
    TN & FP \\
    FN & TP \\
\end{bmatrix}


# **Classification Report**

Classification report memberikan ringkasan yang lebih komprehensif tentang kinerja model dengan menyertakan beberapa metrik seperti precision, recall, f1-score, dan support untuk setiap kelas. Metrik-metrik ini dihitung berdasarkan nilai-nilai dari confusion matrix.

Rumus :    

\begin{align*}
    Precision & = \frac{TP}{TP + FP} \\
    Recall & = \frac{TP}{TP + FN} \\
    F1\text{-}Score & = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}} \\
    Accuracy & = \frac{TP + TN}{TP + TN + FP + FN} \\
\end{align*}


- $TP$ adalah True Positive

- $TN$ adalah True Negative

- $FP$ adalah False Positive

- $FN$ adalah False Negative.

# **Akurasi**

Akurasi adalah ukuran keseluruhan sejauh mana model klasifikasi dapat memprediksi dengan benar. Tujuannya adalah menunjukkan persentase total prediksi yang benar dari semua prediksi yang dilakukan oleh model.

Rumus:
\begin{align*}
Accuracy & = \frac{TP + TN}{TP + TN + FP + FN} \\
\end{align*}

# **Logistic Regression**

Regresi Logistik (Logistic Regression):
Regresi logistik adalah metode statistik yang digunakan untuk memodelkan hubungan antara satu atau lebih variabel independen (fitur) dan variabel dependen biner (dua kelas). Ini digunakan untuk tugas klasifikasi di mana tujuan adalah memprediksi probabilitas keanggotaan dalam salah satu dari dua kelas.

Rumus Logistic Regression:
Fungsi logistik atau sigmoid function digunakan untuk menghasilkan nilai probabilitas antara 0 dan 1. Rumus regresi logistik dapat dituliskan sebagai berikut:

$ P(Y=1) = \frac{1}{1 + e^{-(\beta_0 + \beta_1X_1 + \beta_2X_2 + \ldots + \beta_kX_k)}} $

$ P(Y=0) = 1 - P(Y=1) $



di mana:

$ P(Y=1) $ adalah probabilitas dari kelas 1.

$ P(Y=0) $ adalah probabilitas dari kelas 0.

$ e $ adalah basis dari logaritma natural.

$ \beta_0, \beta_1, \ldots, \beta_k $ adalah parameter model yang perlu diestimasi.

$ X_1, X_2, \ldots, X_k $ adalah variabel independen (fitur).


```python
#membuat objek model regresi logistik dengan menggunakan solver 'newton-cg' dan multi_class 'multinomial' dalam Scikit-learn.
logreg=LogisticRegression(solver='newton-cg',multi_class='multinomial')
```

# **Random Forest**

random forest.PNG

Random Forest adalah algoritma machine learning yang termasuk dalam kategori ensemble learning. Ensemble learning menggabungkan prediksi dari beberapa model untuk meningkatkan performa dan kestabilan model. Random Forest bekerja dengan membuat sejumlah besar decision tree (pohon keputusan) pada data pelatihan dan menggabungkan prediksi dari setiap pohon untuk mendapatkan prediksi akhir.

Beberapa karakteristik Random Forest:

- Setiap pohon dihasilkan secara acak dengan menggunakan subset acak dari data pelatihan.
- Subsample yang digunakan untuk melatih setiap pohon dapat dipilih dengan atau tanpa penggantian (bootstrap).
- Prediksi akhir diperoleh dengan menggabungkan hasil dari setiap pohon, misalnya dengan cara mengambil rata-rata (untuk regresi) atau melakukan voting mayoritas (untuk klasifikasi).

Rumus untuk Prediksi dalam Random Forest :

**klasifikasi :**

$ \hat{Y} = \text{mode}(Y_1, Y_2, \ldots, Y_n) $

di mana $Y_1$ adalah prediksi dari pohon ke-$i$, dan $\hat{Y}$ adalah prediksi akhir.

**regresi :**

$ \hat{Y} = \frac{1}{N} \sum_{i=1}^{N} Y_i $  (untuk regresi)

dimana $Y_1$ adalah prediksi dari pohon ke-$i$, dan $\hat{Y}$ adalah prediksi akhir dan $N$ adalah jumlah pohon.

# **Learning Curve**

Learning Curve adalah grafik yang menunjukkan bagaimana performa model (biasanya dalam hal akurasi atau error) berubah seiring dengan pertambahan jumlah data pelatihan. Kurva pembelajaran membantu kita memahami sejauh mana model kita dapat belajar dari data yang diberikan dan apakah ada manfaat tambahan dalam mengumpulkan lebih banyak data.

**Beberapa Aspek Learning Curve:**

- Underfitting dan Overfitting: Learning curve membantu mengidentifikasi apakah model cenderung underfitting (performa rendah di data pelatihan dan uji) atau overfitting (performa baik di data pelatihan tetapi buruk di data uji).

- Estimasi Kapasitas Model: Jika kurva pembelajaran menunjukkan bahwa akurasi terus meningkat dengan penambahan data, ini mungkin menunjukkan bahwa model memiliki kapasitas untuk terus belajar dari data tambahan.


Learning Curve:

${Performance} \, (Y) \, \text{versus Training Data Size} \, (X) $


```python
# Menghasilkan kurva pembelajaran untuk model RandomForestClassifier
# dengan parameter tertentu pada data latih (X_train, y_train)
# Data latih dibagi menjadi 5 lipatan dalam validasi silang.
lc=learning_curve(RandomForestClassifier(n_estimators=50,criterion='entropy',max_features=6,max_depth=10,random_state=51,
                                             max_leaf_nodes=None,n_jobs=-1,),X_train,y_train,cv=5,n_jobs=-1)

# Menghitung rata-rata skor pelatihan dan pengujian dari hasil kurva pembelajaran
size=lc[0]
train_score=[lc[1][i].mean() for i in range (0,5)]
test_score=[lc[2][i].mean() for i in range (0,5)]

# Menyiapkan dan menampilkan grafik kurva pembelajaran dengan menggunakan matplotlib
fig=plt.figure(figsize=(12,8))
plt.plot(size,train_score)
plt.plot(size,test_score)
```

# **Neural Network**

Jaringan saraf tiruan (Neural Network atau disingkat NN) adalah model matematika yang terinspirasi dari struktur dan fungsi jaringan saraf biologis. Ini terdiri dari neuron buatan (unit komputasi) yang diorganisasikan dalam lapisan. Ada lapisan input, satu atau lebih lapisan tersembunyi, dan lapisan output. Neural network digunakan untuk memodelkan hubungan kompleks dan melakukan tugas seperti klasifikasi, regresi, dan pengenalan pola.


Rumus untuk neuron pada lapisan tersembunyi:

$
z_j^{(l)} = \sum_{i=1}^{n^{(l-1)}} w_{ij}^{(l)} a_i^{(l-1)} + b_j^{(l)}
a_j^{(l)} = f(z_j^{(l)})
$

Rumus untuk neuron pada lapisan output:

$
\hat{y}_k = f(z_k^{(L)})
$

# **Decission Tree**

Decission Tree adalah model prediktif yang digunakan dalam machine learning dan data mining. Model ini memecah data menjadi subset yang lebih kecil berdasarkan keputusan pada fitur-fitur yang dimiliki oleh data. Pada setiap simpul (node) pohon, keputusan dibuat berdasarkan fitur tertentu, dan setiap cabang dari simpul tersebut mengarah ke subset data yang sesuai dengan keputusan tersebut. Pemecahan berlanjut hingga mencapai daun (leaf) pohon, yang mewakili prediksi atau kelas.

Rumus Decission Tree :     

$
\begin{equation}
\begin{split}
& \text{Simpul: } \\
& \text{Decision}(\text{Feature}_i) = \begin{cases}
\text{Leaf}_{1} & \text{if } \text{Feature}_i \leq \theta_{1} \\
\text{Decision}(\text{Feature}_j) & \text{if } \text{Feature}_i > \theta_{1} \text{ and } \text{Feature}_j \leq \theta_{2} \\
\text{Leaf}_{2} & \text{if } \text{Feature}_i > \theta_{1} \text{ and } \text{Feature}_j > \theta_{2} \text{ and } \text{Feature}_k \leq \theta_{3} \\
\vdots \\
\text{Leaf}_{k} & \text{otherwise}
\end{cases}
\end{split}
\end{equation}
 $
